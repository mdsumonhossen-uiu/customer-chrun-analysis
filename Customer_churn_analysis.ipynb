```python
def conv(value):
  if value==1:
    return "yes"
  else:
    return "no"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python
ax=sns.countplot(x="Churn",data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(3,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels=gb.index,autopct= "%1.2f%%")
plt.title("Percentage of Churned Customers",fontsize=10)
plt.show()
gb
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x= "SeniorCitizen", data=df, hue= "Churn")
plt.show()
```

```python
import matplotlib.pyplot as plt
churn_by_senior = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
churn_by_senior_percentage = churn_by_senior.apply(lambda x: x / x.sum(), axis=1)

ax = churn_by_senior_percentage.plot(kind='bar', stacked=True, figsize=(6, 4))
plt.title('Churn Percentage by Senior Citizen Status')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.xticks(rotation=0)

for container in ax.containers:
  labels = [f'{w*100:.1f}%' if (w := v.get_height()) > 0 else '' for v in container]
  ax.bar_label(container, labels=labels, label_type='center')

plt.show()
```
